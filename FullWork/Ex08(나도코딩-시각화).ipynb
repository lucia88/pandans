# 2. 축

```python
X=[1, 2, 3]
Y=[2, 4, 8]
plt.plot(X, Y)
plt.title('꺽은선 그래프', fontdict={"family":"HYGungSo-Bold", "size":"20"})    # 개별 폰트 설정
```

```python
plt.plot(X, Y)
plt.title('꺽은선 그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(X, Y)
plt.title('꺽은선 그래프')
plt.xlabel('X축', color='red')
plt.ylabel('Y축', color='#00AA00')
```

```python
plt.plot(X, Y)
plt.title('꺽은선 그래프')
# plt.xlabel('X축', color='red', loc='right')              # left, center, right
# plt.ylabel('Y축', color='#00AA00' loc='top')             # top, center, bottom 
```

```python
plt.plot(X, Y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```

# 3. 범례 (legend)

```python
plt.plot(X, Y, label="무슨 데이터")
plt.legend()
```

```python
plt.plot(X, Y, label="무슨 데이터")
plt.legend(loc='upper right')             
```

```python
plt.plot(X, Y, label="무슨 데이터")
plt.legend(loc='lower right')     
```

```python
plt.plot(X, Y, label="무슨 데이터")
plt.legend(loc='best')   
```

```python
plt.plot(X, Y, label="범례")
plt.legend(loc=(0.5, 0.5))         # X, Y(0~1)   
```

```python
plt.plot(X, Y, label="범례")
plt.legend(loc=(0.7, 0.8))  
```
